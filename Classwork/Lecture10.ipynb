```python
def generate_histogramavg(x):
  X = genfromtxt(x, delimiter=',')
  n, bins, patches = plt.hist(X,20, facecolor = 'blue', alpha= 0.7)
  plt.show()
  average = mean(X)
  print('2. the average of the dataset is', average)
SAT = genfromtxt('SAT.csv', delimiter=',')
generate_histogramavg('SAT.csv')
LUNCH = genfromtxt('lunch_wait_time.csv', delimiter=',')
generate_histogramavg('lunch_wait_time.csv')
AGE = genfromtxt('student_age.csv', delimiter=',')
generate_histogramavg('student_age.csv')
```

```python
#AGE
#Approximately a uniform distribution therefore p(x) ~ 1/(23-17) = 1/6 for each distinct x
X = genfromtxt('student_age.csv', delimiter=',')
n, bins, patches = plt.hist(X,20, facecolor = 'blue', density = True, alpha= 0.7)
#plot uniform distribution over histogram
plt.plot(bins, np.ones_like(bins)*1/6, linewidth=2, color='r')
plt.show()
average = mean(X)
print('2. the average of the dataset is', average)
```

```python
#SAT
X = genfromtxt('SAT.csv', delimiter=',')
plt.hist(X,30, facecolor = 'blue', density = True, alpha= 0.7)
#approximately a gaussian distribution so p(x) ~ (1/(std*np.sqrt(2*np.pi))) * np.exp(-((x-avg)**2/2*std**2))
x = np.linspace(np.mean(X)-4*np.std(X), np.mean(X)+4*np.std(X), 100)
y = (1 / (np.std(X) * np.sqrt(2 * np.pi))) * np.exp(-(x - np.mean(X)) ** 2 / (2 * np.std(X) ** 2))
plt.plot(x,y, linewidth=2, color='r')
plt.show()
average = mean(X)

plt.xlabel = 'SAT'
plt.ylabel = 'Frequency'
plt.title = 'Histogram of SAT'
```
