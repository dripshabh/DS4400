```python
def closed_form(A,y):
  return np.linalg.inv(A.T.dot(A)).dot(A.T).dot(y)

A = np.array([[0,0,2],
              [0,2,-1],
              [2,-1,0]])
y = np.array([[-3,0,2]]).T
print(closed_form(A,y))
```

```python
def obj(x):
  return 2*x**2 + 1
def g(x):
  return -(x - 1)
constraints = [{'type': 'ineq', 'fun': g}]
result = scipy.optimize.minimize(obj, -2, constraints = constraints)
print(result)
```

```python
def obj(x):
  return x[0] + 4 * x[1]**2
def h(x):
  return x[0]**2 - 5*x[0] + 4
def g(x):
  return -(1 - x[1])
constraints = [{'type': 'ineq', 'fun': g}, {'type': 'eq', 'fun': h}]
result = scipy.optimize.minimize(obj, np.array([2,2]).T, constraints = constraints)
print(result)
```

```python
def obj(x):
  return 1/2*x[0]**2 + x[1]**2
def g1(x):
  return (x[1] - 1)
def g2(x):
  return (x[0]-3*x[1])
constraints = [{'type': 'ineq', 'fun': g1}, {'type': 'ineq', 'fun': g2}]
result = scipy.optimize.minimize(obj, np.array([2,2]).T, constraints = constraints)
print(result)
```
