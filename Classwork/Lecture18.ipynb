```python
def calc_entropy(data):
    """
    Calculate the entropy of the dataset based on the class labels.
    """
    labels = data[:, -1]
    label_counts = Counter(labels)
    total_samples = len(labels)

    # Calculate entropy
    entropy = -sum((count / total_samples) * np.log2(count / total_samples)
                   for count in label_counts.values())

    return entropy
calc_entropy(data)
```
