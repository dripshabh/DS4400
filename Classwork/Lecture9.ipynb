```python
import numpy as np
from numpy import ones
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn import preprocessing
from numpy import genfromtxt
from numpy import mean
from numpy.random import randn
from scipy.integrate import quad
import sys
np.set_printoptions(precision=4)
np.set_printoptions(threshold=30)
np.set_printoptions(linewidth=300)
np.set_printoptions(suppress=True)
np.set_printoptions(threshold=sys.maxsize)
```

```python
#data
X = np.array([[0, 1, 2, 3]]).T
Y = np.array([[0, 1, 1, 2]]).T
w0 = np.array([[0, 0, 1]]).T
phi = np.array([[0, 1, 4, 9],
                [0, 1, 2, 3],
                [1, 1, 1, 1]]).T
lmbda = 0.2
eta = 0.01

def calculateMSE(w, phi, y):
  n = X.size
  f0 = 0
  for phii, yi in zip (phi,y):
    phii = np.reshape(phii,(-1,1))
    f0 += (w.T.dot(phii) - yi)**2
  return (f0/n).item() + lmbda*np.sum(np.abs(w))

def f_grad(w, phi, y):
  n = X.size
  return 2/n*phi.T.dot(phi.dot(w) - y) + lmbda*np.sign(w)


mse_list = []
for i in range(1000):
  mse_list.append(calculateMSE(w, phi, Y))
  w = w - eta*f_grad(w, phi, Y)

plt.plot(mse_list)
plt.title('MSE over GD')
plt.xlabel('steps')
plt.ylabel('MSE')
plt.show()
```
