```python
import numpy as np
import matplotlib.pyplot as plt

# Data points
x = np.array([[1, 2, 1.5, 3]]).T
y = np.array([[1, 1, 0, 2]]).T

# Quadratic feature map
phi = np.array([[1, 1, 1],
                     [4,2,1],
                     [2.25,1.5,1],
                     [9,3,1]])

# Cubic feature map
phi2 = np.array([[1, 1, 1, 1],
                     [8, 4,2,1],
                     [3.375, 2.25,1.5,1],
                     [27, 9,3,1]])
# Quartic feature map
phi3 = np.array([[1, 1, 1, 1, 1],
                     [16, 8, 4,2,1],
                     [5.0625, 3.375, 2.25,1.5,1],
                     [81, 27, 9,3,1]])
# Quintic feature map
phi4 = np.array([[1, 1, 1, 1, 1,1],
                     [32, 16, 8, 4,2,1],
                     [7.59375, 5.0625, 3.375, 2.25,1.5,1],
                     [243, 81, 27, 9,3,1]])
# Function to calculate w
def findw(phi, y):
    return np.linalg.inv(phi.T @ phi) @ (phi.T @ y)

# Compute weights
w = findw(phi, y)   # Quadratic weights
w2 = findw(phi2, y) # Cubic weights
w3 = findw(phi3, y) # Quartic weights
w4 = findw(phi4, y) # Quintic weights

# Generate data for the fitted functions
x_plot = np.linspace(min(x), max(x), 100).reshape(-1, 1)  # Smooth range of x values
phi_plot = np.hstack([x_plot**2, x_plot, np.ones_like(x_plot)])  # Quadratic feature map for plotting
phi2_plot = np.hstack([x_plot**3, x_plot**2, x_plot, np.ones_like(x_plot)])  # Cubic feature map for plotting
phi3_plot = np.hstack([x_plot**4, x_plot**3, x_plot**2, x_plot, np.ones_like(x_plot)])  # Quartic feature map for plotting
phi4_plot = np.hstack([x_plot**5, x_plot**4, x_plot**3, x_plot**2, x_plot, np.ones_like(x_plot)])  # Quintic feature map for plotting

# Predicted values
y_fit = phi_plot @ w   # Quadratic fit
y2_fit = phi2_plot @ w2  # Cubic fit
y3_fit = phi3_plot @ w3  # Quartic fit
y4_fit = phi4_plot @ w4  # Quintic fit

# Plotting
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='red', label='Data points')  # Plot the original x, y points
plt.plot(x_plot, y_fit, color='blue', label='Quadratic Fit')  # Quadratic fit
plt.plot(x_plot, y2_fit, color='green', label='Cubic Fit')  # Cubic fit
plt.plot(x_plot, y3_fit, color='purple', label='Quartic Fit')  # Quartic fit
plt.plot(x_plot, y4_fit, color='orange', label='Quintic Fit')  # Quintic fit
plt.title('Fitted Functions and Data Points')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
from numpy import ones
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn import preprocessing
from numpy import genfromtxt
from numpy import mean
from numpy.random import randn
import sys
np.set_printoptions(precision=4)
np.set_printoptions(threshold=30)
np.set_printoptions(linewidth=300)
np.set_printoptions(suppress=True)
np.set_printoptions(threshold=sys.maxsize)
```

```python
X = genfromtxt('stock_prediction_data_scaled.csv', delimiter=',')
y = genfromtxt('stock_price.csv', delimiter=',')
y = y.reshape(y.size,1)
X = preprocessing.scale(X)
n = X.shape[0]
print (X)
print (y)
```

```python
#Linear Regression closed form


phi = np.hstack((X,ones((n,1))))
d = phi.shape[1]
w = randn(d,1)
def findw(phi, y):
  return np.linalg.inv((phi.T.dot(phi))).dot((phi.T).dot(y))
closed_form_w = findw(phi, y)
print(closed_form_w)
#print out predictions vs actual
print(phi.dot(closed_form_w),y)


#MSE
def calculateMSE(w):
  f0 = 0
  for phii, yi in zip (phi,y):
    phii = np.reshape(phii,(d,1))
    f0 += (w.T.dot(phii) - yi)**2
  return (f0/n).item()
print(calculateMSE(closed_form_w))
```

```python
def f_grad(w):
  advanced_approach = False
  if advanced_approach:
    return 2/n*phi.T.dot(phi.dot(w) - y)
  else:
    f0 = np.zeros((d,1))
    for phii,yi in zip(phi,y):
      phii = np.reshape(phii,(d,1))
      f0 += (w.T.dot(phii) - yi)*phii
  return 2/n*f0
f_grad(closed_form_w)
```

```python
mse_list = []
for i in range(200):
  w -= 0.01*f_grad(w)
  mse_list.append(calculateMSE(w))
```

```python
plt.plot(mse_list)
plt.title('MSE over GD')
plt.xlabel('steps')
plt.ylabel('MSE')
plt.show()
```
